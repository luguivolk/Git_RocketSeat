```python
# ============================================
# SISTEMA DE LISTA DE COMPRAS
# Desafio pr√°tico - Curso Data Analytics
# ============================================

# Lista global para armazenar os itens
itens = []

def adicionar_item():
    """Adiciona um novo item √† lista de compras"""
    print("\n=== ADICIONAR ITEM ===")

    # Valida√ß√£o do nome
    while True:
        nome = input("Nome do item: ").strip()
        if nome:
            break
        print("‚ùå O nome n√£o pode ser vazio!")

    # Valida√ß√£o da quantidade
    while True:
        try:
            quantidade = int(input("Quantidade: "))
            if quantidade > 0:
                break
            print("‚ùå A quantidade deve ser maior que zero!")
        except ValueError:
            print("‚ùå Digite um n√∫mero v√°lido!")

    # Criar o item
    item = {
        'nome': nome,
        'quantidade': quantidade,
        'comprado': False
    }

    itens.append(item)
    print(f"‚úÖ Item '{nome}' adicionado com sucesso!")


def listar_itens():
    """Exibe todos os itens da lista"""
    print("\n=== LISTA DE COMPRAS ===")

    if not itens:
        print("üì≠ Lista vazia!")
        return

    for i, item in enumerate(itens, 1):
        status = "‚úîÔ∏è Comprado" if item['comprado'] else "‚è≥ Pendente"
        print(f"{i}. {item['nome']} - Qtd: {item['quantidade']} - {status}")


def marcar_comprado():
    """Marca um item como comprado"""
    listar_itens()

    if not itens:
        return

    print("\n=== MARCAR COMO COMPRADO ===")

    while True:
        try:
            indice = int(input("N√∫mero do item: ")) - 1
            if 0 <= indice < len(itens):
                itens[indice]['comprado'] = True
                print(f"‚úÖ '{itens[indice]['nome']}' marcado como comprado!")
                break
            print("‚ùå N√∫mero inv√°lido!")
        except ValueError:
            print("‚ùå Digite um n√∫mero v√°lido!")


def remover_item():
    """Remove um item da lista"""
    listar_itens()

    if not itens:
        return

    print("\n=== REMOVER ITEM ===")

    while True:
        try:
            indice = int(input("N√∫mero do item: ")) - 1
            if 0 <= indice < len(itens):
                item_removido = itens.pop(indice)
                print(f"üóëÔ∏è '{item_removido['nome']}' removido!")
                break
            print("‚ùå N√∫mero inv√°lido!")
        except ValueError:
            print("‚ùå Digite um n√∫mero v√°lido!")


# ============================================
# DESAFIOS EXTRAS (OPCIONAIS)
# ============================================

def editar_quantidade():
    """Edita a quantidade de um item"""
    listar_itens()

    if not itens:
        return

    print("\n=== EDITAR QUANTIDADE ===")

    while True:
        try:
            indice = int(input("N√∫mero do item: ")) - 1
            if 0 <= indice < len(itens):
                nova_qtd = int(input("Nova quantidade: "))
                if nova_qtd > 0:
                    itens[indice]['quantidade'] = nova_qtd
                    print(f"‚úÖ Quantidade atualizada!")
                    break
                print("‚ùå Quantidade deve ser maior que zero!")
        except ValueError:
            print("‚ùå Digite um n√∫mero v√°lido!")


def listar_pendentes():
    """Mostra apenas itens n√£o comprados"""
    print("\n=== ITENS PENDENTES ===")

    pendentes = [item for item in itens if not item['comprado']]

    if not pendentes:
        print("üéâ Tudo comprado!")
        return

    for i, item in enumerate(pendentes, 1):
        print(f"{i}. {item['nome']} - Qtd: {item['quantidade']}")


# ============================================
# FUN√á√ÉO PRINCIPAL - MENU
# ============================================

def main():
    """Fun√ß√£o principal com menu interativo"""

    while True:
        print("\n" + "="*40)
        print("üõí  LISTA DE COMPRAS")
        print("="*40)
        print("1. Adicionar item")
        print("2. Listar itens")
        print("3. Marcar item como comprado")
        print("4. Remover item")
        print("5. Editar quantidade (extra)")
        print("6. Listar apenas pendentes (extra)")
        print("7. Sair")
        print("="*40)

        opc = input("Escolha uma op√ß√£o: ").strip()

        if opc == '1':
            adicionar_item()
        elif opc == '2':
            listar_itens()
        elif opc == '3':
            marcar_comprado()
        elif opc == '4':
            remover_item()
        elif opc == '5':
            editar_quantidade()
        elif opc == '6':
            listar_pendentes()
        elif opc == '7':
            print("\nüëã At√© logo!")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida!")


# ============================================
# EXECUTAR O PROGRAMA
# ============================================

if __name__ == "__main__":
    main()
```
